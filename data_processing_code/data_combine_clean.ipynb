```python
with open("business_with_reviews.json") as f:
    business = json.load(f)   # <-- data is already list[dict]

business = pd.DataFrame(business)
business.to_parquet("business.parquet")
```

```python
business_map = {}

for datapoint in business:
  del datapoint['sample_reviews']
  id = datapoint['business_id']
  del datapoint['business_id']
  business_map[id] = datapoint
```

```python
business_map
```

```python
def process_reviews(row):
    reviews = row['reviews']
    processed_reviews = []
    for review in reviews:
      adding = {}
      if review['business_id'] in business_map:
        info = business_map[review['business_id']]
        for key in info:
          adding[key] = info[key]
        for key in review:
          adding[key] = review[key]
        processed_reviews.append(adding)
    return processed_reviews
```

```python
merged['reviews'] = merged.apply(lambda row: process_reviews(row), axis=1)
```

```python
def removed_nulls(row):
  reviews = row['reviews']
  new_reviews = []
  for review in reviews:
    att_key = list(review['attributes'].keys()) if review['attributes'] is not None else []
    for k in att_key:
      if review['attributes'][k] is None:
        del review['attributes'][k]
    if att_key == []:
      del review['attributes']
    new_reviews.append(review)
  return new_reviews
```

```python
for i in range(5):
  data = pd.read_parquet(f"user_data_part{i}.parquet")
  data['reviews'] = data.apply(lambda row: removed_nulls(row), axis=1)
  data.to_parquet(f"user_data_part{i}.parquet")
  print(f"Finish - {i}")
```
